---

### Treatments.csv

#### Table Purpose
This table records medical treatments provided during patient appointments.

#### Columns Explanation

- **treatment_id**: Unique identifier for each treatment record.
- **appointment_id**: Foreign key linking the treatment to a specific appointment.
- **treatment_type**: Category or type of medical treatment provided.
- **description**: Detailed explanation of the treatment performed.
- **cost**: Financial cost associated with the treatment.
- **treatment_date**: Date when the treatment was administered.

---

### Appointments.csv

#### Table Purpose
This table tracks scheduled and completed appointments between patients and doctors.

#### Columns Explanation

- **appointment_id**: Unique identifier for each appointment.
- **patient_id**: Foreign key referencing the patient attending the appointment.
- **doctor_id**: Foreign key referencing the doctor assigned to the appointment.
- **appointment_date**: Scheduled date of the appointment.
- **appointment_time**: Scheduled time of the appointment.
- **reason_for_visit**: Stated reason or medical concern for the visit.
- **status**: Current status of the appointment (e.g., Completed, Cancelled, No-Show, Scheduled).

---

### Billing.csv

#### Table Purpose
This table stores financial transaction records related to patient treatments and billing operations.

#### Columns Explanation

- **bill_id**: Unique identifier for each billing transaction.
- **patient_id**: Foreign key referencing the patient associated with the bill.
- **treatment_id**: Foreign key linking the bill to a specific treatment provided.
- **bill_date**: Date when the bill was generated.
- **amount**: Total monetary value charged for the treatment (in hospital currency).
- **payment_method**: Method used for payment (e.g., Cash, Credit Card, Insurance).
- **payment_status**: Status of the payment (e.g., Paid, Pending, Cancelled).

---

### Doctors.csv

#### Table Purpose
This table contains detailed information about doctors working within the hospital system.

#### Columns Explanation

- **doctor_id**: Unique identifier for each doctor.
- **first_name**: Doctor’s first name.
- **last_name**: Doctor’s last name.
- **specialization**: Medical specialty (e.g., Cardiology, Neurology, Orthopedics).
- **phone_number**: Doctor’s contact number.
- **years_experience**: Total years of professional medical experience.
- **hospital_branch**: Branch or location where the doctor practices.
- **email**: Official email address of the doctor.

---

### Patients.csv

#### Table Purpose
This table stores demographic and registration information of hospital patients.

#### Columns Explanation

- **patient_id**: Unique identifier for each patient.
- **first_name**: Patient’s first name.
- **last_name**: Patient’s last name.
- **gender**: Patient’s gender.
- **date_of_birth**: Patient’s birth date (used to calculate age).
- **contact_number**: Patient’s primary phone number.
- **address**: Residential address of the patient.
- **registration_date**: Date when the patient was registered in the hospital system.
- **insurance_provider**: Name of the patient’s insurance company.
- **insurance_number**: Unique insurance policy number.
- **email**: Patient’s email address.

```python
#Before we head to the Next Step the Date Columns data type need to be changed

folder_path = 'D:\Data Analysist\EraaSoft Data Analysis Diploma\Projects\Final Project (Graduation)\Database'
column_names = ['appointment_date','bill_date','date_of_birth','registration_date','treatment_date']

# Use glob to find all CSV files in the specified folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))

# Loop over all CSV files
for file in csv_files:
    print(f"Processing file: {file}")
    
    try:
        # Read CSV file
        df = pd.read_csv(file)
        
        # Loop through each column in the list
        for col in column_names:
            if col in df.columns:
                print(f"  Converting column: {col}")
                df[col] = pd.to_datetime(df[col], errors='coerce')
                
            else:
                print(f"  Column not found: {col}")

        # Save the modified DataFrame back to the same file
        df.to_csv(file, index=False)
        print(f"Successfully updated and saved: {file}\n")
        
    except Exception as e:
        print(f"Error processing file {file}: {e}\n")

print("All files processed.")
```

**The Next Setp will be Transfrom the Dataframes To a Cleaned CSV Files and apply it to a database Using SQL Server Managment Tool**

```python
# Define The Target folder
folder_path = r"D:\Data Analysist\EraaSoft Data Analysis Diploma\Projects\Final Project (Graduation)\Database"

# Get all CSV files inside the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))


# New folder for cleaned files
cleaned_folder = os.path.join(folder_path, "Cleaned Version")
# Create cleaned folder if it does not exist
os.makedirs(cleaned_folder, exist_ok=True)


for file_path in csv_files:
    df = pd.read_csv(file_path)
    
    # Extract filename without extension
    file_name = os.path.splitext(os.path.basename(file_path))[0]    
    # Create new cleaned file name
    cleaned_file_name = f"{file_name}_Cleaned.csv" 
   
    # Full Cleaned path
    cleaned_path = os.path.join(cleaned_folder, cleaned_file_name)
    
    # Export DataFrame again as CSV
    df.to_csv(cleaned_path, index=False)
    
    # Print info
    print(f"Original File: {file_name}.csv")
    print(f"Saved Cleaned File: {cleaned_file_name}")
    print("-" * 50)
```
